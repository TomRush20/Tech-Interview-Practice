## 1. Introduction to Matplotlib
- What is Matplotlib?
Matplotlib is a data visualization library in python. It is a common library data analysts use to graph data.

- Brief history and purpose:
It was released in 2003 and is used for programmers to easily visualize their data within their code

- Overview of the Matplotlib library:
In this notebook I will be covering how to install and use matplotlib as well as how to make a line plot, scatter plot, histogram, and bar plot. I will also be going over how to customize each plot.

- Installing Matplotlib:

- Importing Matplotlib in Python:
When you import matplotlib use the statement import matplotlib.pyplot as plt. Also you probably want to import pandas as pd

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#The first dataset we will be using is a dataset containing statistics on COVID-19. Source: Our World In Data
covid_df = pd.read_csv('Data/owid-covid-data.csv')
covid_df_usa = covid_df[covid_df['iso_code'] == 'USA']
covid_df_usa['date'] = pd.to_datetime(covid_df_usa['date'])
covid_df_usa_march = covid_df_usa[covid_df_usa['date'] >= '2020-03-01']
covid_df_usa_march = covid_df_usa_march[covid_df_usa_march['date'] <= '2020-03-31']
covid_df_usa_march.head()
```

## 3. Basic Plotting with Matplotlib
- Create a simple line plot using plt.plot()
- For this example, we will be using a covid dataset to plot number of covid cases in March over time in the United States

```python
#The x-axis column goes first and the y_axis column goes second
plt.plot(covid_df_usa_march['date'], covid_df_usa_march['total_cases'])

#Creating a title for the graph
plt.title('USA Covid Cases in March')

#Labeling the x and y axis
plt.xlabel('Date')
plt.ylabel('COVID-19 Cases')

#Rotate the x axis labels
plt.xticks(rotation=35)
plt.show()
```

- Scatter plots
- Create a scatterplot using plt.scatter(xdata, ydata)
- For this scatterplot we will be using the iris dataset which compares flower petal sizes.

```python
iris_df = pd.read_csv('Data/iris.csv')
iris_df.head()
```

```python
#Create a scatter plot comparing the sepal_length and petal_length of each flower
#You can change the marker size by changing the s parameter
plt.scatter(iris_df['sepal_length'], iris_df['petal_length'], s=50)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```
