## 4. More Plots
- Bar Charts
- Create a simple bar chart using plt.bar()


Parameters:


x: The locations on the x-axis where the bars will be placed. This can be a list or an array of values.

height: The heights of the bars. This can also be a list or an array corresponding to each location in x.

width: (optional) The width of the bars. Default is 0.8. This controls how wide each bar is on the x-axis.

bottom: (optional) The y-coordinate(s) of the bars' bottom edges. Default is 0, which means the bars start from the x-axis.

align: (optional) Alignment of the bars. Can be 'center' (default) or 'edge'. If 'center', bars are centered on the x locations; if 'edge', they align with the edges of the specified x locations.

color: (optional) The color(s) of the bars. Can be a single color or a list of colors for each bar.

edgecolor: (optional) The color of the bar edges. Useful for adding a border around bars.

label: (optional) A string that can be used to label the bars in the legend.

tick_label: (optional) Labels for the ticks on the x-axis. Useful for providing more descriptive labels for each bar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris_df = pd.read_csv('Data/iris.csv')
iris_df['species'].value_counts()
```

```python
#We will create a simple bar chart that compares the average petal_length of each flower in the iris dataset
iris_virginica = iris_df[iris_df['species'] == 'virginica']
iris_versicolor = iris_df[iris_df['species'] == 'versicolor']
iris_setosa = iris_df[iris_df['species'] == 'setosa']

avg_virginica = np.mean(iris_virginica['petal_length'])
avg_versicolor = np.mean(iris_versicolor['petal_length'])
avg_setosa = np.mean(iris_setosa['petal_length'])

species = ['setosa', 'versicolor', 'virginica']
avg_petal_length = [avg_setosa, avg_versicolor, avg_virginica]

#First parameter is the x axis and second parameter is height, you can use any of the optional parameters listed above to 
#customize the graph
plt.bar(species, avg_petal_length)
plt.title('Avg Petal Length per Species')
plt.ylabel('Petal Length')
plt.show()
```

- Horizontal Bar Charts
- These are the same as the vertical bar chart above, you just have to use plt.barh()


Parameters:

y: The locations on the y-axis where the bars will be placed. This can be a list or an array of values.

width: The width of the bars. This can also be a list or an array of values corresponding to each location in y.

height: (optional) The height of the bars. Default is 0.8, which controls how tall each bar is on the y-axis.

left: (optional) The x-coordinate(s) of the bars' left edges. Default is 0, meaning the bars start from the origin.

align: (optional) Alignment of the bars. Can be 'center' (default) or 'edge'. If 'center', bars are centered on the y locations; if 'edge', they align with the edges of the specified y locations.

color: (optional) The color(s) of the bars. Can be a single color or a list of colors for each bar.

edgecolor: (optional) The color of the bar edges. Useful for adding a border around bars.

label: (optional) A string that can be used to label the bars in the legend.

tick_label: (optional) Labels for the ticks on the y-axis. Useful for providing more descriptive labels for each bar.

```python
plt.barh(species, avg_petal_length)
plt.title('Avg Petal Length per Species')
plt.xlabel('Petal Length')
plt.show()
```

- Histograms


plt.hist() creates a histogram.


I am going to create a histogram that shows the distribution of a random set of numbers. 


Parameters: 


x: The input data (array-like) to be binned into the histogram.

bins:

Can be an integer, specifying the number of equal-width bins in the range.
Can be a sequence, specifying the bin edges.
range: (optional) A tuple specifying the lower and upper range of the bins. If not provided, the range is determined from the data.

density: (optional) If True, the histogram is normalized to form a probability density, meaning the area under the histogram will sum to 1.

weights: (optional) An array of the same shape as x that defines the weights for each bin.

color: (optional) Color of the bars. This can be a single color or a list of colors.

alpha: (optional) The transparency level of the bars (0 is fully transparent and 1 is fully opaque).

edgecolor: (optional) Color of the edges of the bars.

label: (optional) A string to label the dataset, useful for legends.

orientation: (optional) If set to 'horizontal', the histogram is drawn horizontally.

histtype: (optional) The type of histogram to draw. Options include:

'bar': Standard bar-type histogram.
'barstacked': Stacked bar-type histogram.
'step': Line histogram.
'stepfilled': Filled line histogram.
linestyle: (optional) Line style for the edges of the bars (useful for histtype='step').

```python
#1000 random integers from a standard normal distribution that we can use to make our histogram
data = np.random.randn(1000) 

plt.hist(data, bins=30) #histogram with 30 equal width bars
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#We can normalize the distribution by adding the density parameter

plt.hist(data, bins=30, density=True) #histogram with 30 equal width bars
plt.title('Normalized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#We can also create multiple histograms on the same plot
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1  # Shifted distribution

#This creates two distributions from both datasets
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1', color='blue')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Multiple Histograms')
#This creates a legend in the graph with labels corresponding to the label parameter
plt.legend()
plt.show()
```
