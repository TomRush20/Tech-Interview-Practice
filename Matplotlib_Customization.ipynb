## 5. Customizing Plots

- Matplotlib plots can be customized with annotations, legends, and text

plt.annotate() creates an annotation within the graph.


Essential Parameters:


text: The text of the annotation.


xy: The point (x, y) to annotate.


xytext: The position (x, y) to place the text at. If None, defaults to xy.

Lets look back at the bar plot we had in the previous notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

iris_df = pd.read_csv('Data/iris.csv')
```

```python
#We will create a simple bar chart that compares the average petal_length of each flower in the iris dataset
iris_virginica = iris_df[iris_df['species'] == 'virginica']
iris_versicolor = iris_df[iris_df['species'] == 'versicolor']
iris_setosa = iris_df[iris_df['species'] == 'setosa']

avg_virginica = np.mean(iris_virginica['petal_length'])
avg_versicolor = np.mean(iris_versicolor['petal_length'])
avg_setosa = np.mean(iris_setosa['petal_length'])

species = ['setosa', 'versicolor', 'virginica']
avg_petal_length = [avg_setosa, avg_versicolor, avg_virginica]

#First parameter is the x axis and second parameter is height, you can use any of the optional parameters listed above to 
#customize the graph
plt.bar(species, avg_petal_length)
plt.title('Avg Petal Length per Species')
plt.ylabel('Petal Length')

#Add an annotation to the graph for the point, the ha parameter aligns the text in the center of the point being annotated.
plt.annotate('Lowest petal length', (0,2), ha='center')

plt.show()
```

- plt.annotate() can also create an arrow to point to a certain spot on the graph.

```python
#First parameter is the x axis and second parameter is height, you can use any of the optional parameters listed above to 
#customize the graph
plt.bar(species, avg_petal_length)
plt.title('Avg Petal Length per Species')
plt.ylabel('Petal Length')

#Add an annotation to the graph for the point, the ha parameter aligns the text in the center of the point being annotated.
plt.annotate('Lowest petal length', (0,2), ha='center')

#This places an annotation with a red arrow pointing at (1,4.35) on the graph
plt.annotate('Middle petal length', xy=(1,4.35), xytext=(1,5), ha='center', arrowprops=dict(facecolor='red'))

plt.show()
```

- Adjusting the size of each graph.
- The figure of each graph is how much space the graph has.
- You can change the size of each graph by adding the line plt.figure(figsize=(width, length)) to the beginning of the code cell.

```python
#This is a 15x10 figure that holds the bar plot
plt.figure(figsize=(15,10))
#First parameter is the x axis and second parameter is height, you can use any of the optional parameters listed above to 
#customize the graph
plt.bar(species, avg_petal_length)
plt.title('Avg Petal Length per Species')
plt.ylabel('Petal Length')

#Add an annotation to the graph for the point, the ha parameter aligns the text in the center of the point being annotated.
plt.annotate('Lowest petal length', (0,2), ha='center')

#This places an annotation with a red arrow pointing at (1,4.35) on the graph
plt.annotate('Middle petal length', xy=(1,4.35), xytext=(1,5), ha='center', arrowprops=dict(facecolor='red'))

plt.show()
```

- Creating multiple plots on one figure.
- you can use plt.subplots(#rows, #cols) to create this.

```python
#We are going to use subplots to put our bar graph next to our histogram. We are using a 2 row 3 column matrix
fig, axs = plt.subplots(2, 3, figsize=(10,8))

#axs[0,0] is the first row and the first column of the subplot
axs[0,0].bar(species, avg_petal_length)

#axs[0,1] is the first row and second column of the subplot
#1000 random integers from a standard normal distribution that we can use to make our histogram
data = np.random.randn(1000) 

axs[0,1].hist(data, bins=30) #histogram with 30 equal width bars

plt.show()
```
